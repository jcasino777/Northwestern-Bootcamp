### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data['SN'].unique())
player_count
```

# Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Analysis_df= pd.DataFrame([len(purchase_data['Item Name'].unique())])
Analysis_df.columns = ['Number of Unique Items']
Analysis_df['Avg Price'] = purchase_data['Price'].mean()
Analysis_df['Number of Purchases'] = purchase_data['Purchase ID'].count()
Analysis_df['Total Revenue'] = purchase_data['Price'].sum()
Analysis_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
GenderDemographics = pd.DataFrame([purchase_data.groupby('Gender').count()['SN']]).T
GenderDemographics = GenderDemographics.rename(columns={"SN": "Total Count"})
total = GenderDemographics['Total Count'].sum()
GenderDemographics['% of Players'] = GenderDemographics['Total Count'] / total * 100
GenderDemographics.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchaseAnalysis = pd.DataFrame([purchase_data.groupby('Gender').count()['SN']]).T
GenderDemographics = GenderDemographics.rename(columns={"SN": "Total Count"})
PurchaseAnalysis['Avg Purchase Price'] = purchase_data.groupby('Gender').mean()['Price']
PurchaseAnalysis['Total Purchase Value'] = purchase_data.groupby('Gender').sum()['Price']
PurchaseAnalysis.head()
```

## Age Demographics


* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
heroes_bins = [0,10, 15, 20, 25, 30, 35, 40,200]
heroes_labels = ["<10", "10-14", "15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data['Age Group'] = pd.cut(purchase_data["Age"], heroes_bins, labels=heroes_labels)
AgeDemographic = pd.DataFrame([purchase_data.groupby('Age Group').count()['SN']]).T
AgeDemographic = AgeDemographic.rename(columns={"SN": "Total Count"})
total = AgeDemographic['Total Count'].sum()
AgeDemographic['% of Players'] = AgeDemographic['Total Count'] / total * 100
AgeDemographic.head()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PurchaseAnalysisAge = pd.DataFrame([purchase_data.groupby('Age Group').count()['SN']]).T
PurchaseAnalysisAge = PurchaseAnalysisAge.rename(columns={"SN": "Total Count"})
PurchaseAnalysisAge['Avg Purchase Price'] = purchase_data.groupby('Age Group').mean()['Price']
PurchaseAnalysisAge['Total Purchase Value'] = purchase_data.groupby('Age Group').sum()['Price']
PurchaseAnalysisAge.head()
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
TopSpenders = pd.DataFrame([purchase_data.groupby('SN').count()['Purchase ID']]).T
TopSpenders = TopSpenders.rename(columns={"Purchase ID": "Purchase Count"})
TopSpenders['Avg Purchase Price'] = purchase_data.groupby('SN').mean()['Price']
TopSpenders['Total Purchase Value'] = purchase_data.groupby('SN').sum()['Price']
SortedTopSpenders = TopSpenders.sort_values("Purchase Count",ascending=False)
SortedTopSpenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
reduced = pd.DataFrame(purchase_data[['Item ID','Item Name','Price','SN']])
reduced = reduced.sort_values("Item ID",ascending=False)
TopItems = pd.DataFrame(reduced.groupby(['Item ID','Item Name','Price']).count())
TopItems = TopItems.rename(columns={"SN": "Purchase Count"})
TopItems.reset_index(inplace=True)  
TopItems.set_index('Item ID')
TopItems.head(50)
# prices = purchase_data[['Item ID','Price']]
# prices = prices.set_index('Item ID')
# merged_TopItems = pd.merge(TopItems, prices, on="Item ID",how="left")
# TopItems.head()
# prices.head()
# merged_TopItems.head()
TopItems['Total Purchase Value'] = TopItems['Purchase Count'] * TopItems['Price']
TopItems.sort_values("Purchase Count",ascending=False)
```
